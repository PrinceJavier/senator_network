```python
# convert senators from strings to list
import ast
props = df0.proponent.values # get proponents string
props = [re.sub("(?<=\').* by ", "", i) for i in props]
props = [i.replace(".", "") for i in  props] # remove period
props = [i.replace("\"", "") for i in  props] # remove "
# replace names for splitting
props = [i.replace("joel,", "joel#") for i in  props]
props = [i.replace("bam,", "bam#") for i in  props]
props = [i.replace("risa,", "risa#") for i in  props]
# props = [re.sub(".* by ", "", i) for i in props]

# props = [i.split("#") for i in  props]
props = [ast.literal_eval(props[i]) for i in range(len(props))] # convert to list

props2 = []
for senators in props:
    senators_new = []
    for senator in senators:
        senator_new = senator.strip()
        senators_new.append(senator_new)
    props2.append(senators_new)

props3 = []
for senators in props2:
    senators_new = []
    for senator in senators:
        senator_split = senator.split("#")
        for each in senator_split:
            senators_new.append(each.strip())
    props3.append(senators_new)
```

```python
# simulating filtering

def filter_return_props(df, filter_string):
    if filter_string == "all":
            df_filtered = df[:]
    else:
        df_filtered = df[df[filter_string] == 1]

    props = df_filtered.proponent.values
    props = [props[i] for i in range(len(props))] # convert to list
    return props
```

```python
import ast

def unique_elements(element_set, truncate=False):
    if truncate == True:
        unique_elems = list(set([item.split(',')[0].replace(",", "") for sublist in element_set for item in ast.literal_eval(sublist)]))
    else:
        unique_elems = list(set([item for sublist in element_set for item in ast.literal_eval(sublist)]))
    return unique_elems
```

```python
def make_matrix(element_set, unique_elements, include_diag=False):
    # making adjacency matrix based on processed dataset [[a, v], [a, c]]

    matrix = np.zeros((len(unique_elements), len(unique_elements)))

    for i in range(len(unique_elements)):
        for j in range(len(unique_elements)):
            for x in element_set:
                if include_diag == True:
                    # including filed by same 1 proppnent
                    if i == j:
                        x = ast.literal_eval(x)
                        if len(x) == 1 and unique_elements[i] in x and unique_elements[j] in x:
                            matrix[i][j] += 1
                    if i != j:                
                        if unique_elements[i] in x and unique_elements[j] in x:
                            matrix[i][j] += 1
                else:
                    if i != j:                
                        if unique_elements[i] in x and unique_elements[j] in x:
                            matrix[i][j] += 1
    return matrix
```

```python
elems = filter_return_props(df1, "all")
unique_elems = unique_elements(elems)
matrix = make_matrix(elems, unique_elems, include_diag = False)
matrix
np.sum(matrix, axis=0), np.sum(matrix, axis=1) # no. of bills filed per senator
```

```python
from IPython.display import display, IFrame, HTML
import os


def show_app(app, port=9999, width=900, height=700):
    '''
    Accept app - a dash.Dash() data
    Return a running dash data running on the server
    '''

    # we set the server (host)
    host = 'localhost'

    # we set the URL of the page as localhost:9999
    url = f'http://{host}:{port}'  # use f to pass the variable in {}

    # display in HTML an open in new tab link
    display(HTML(f"<a href='{url}' target='_blank'>Open in a new tab </a>"))

    # display url in IFrame
    display(IFrame(url, width=width, height=height))

    # ???
    app.css.config.serve_locally = True
    app.scripts.config.serve_locally = True

    # Return app = dash.Dash() running server
    return app.run_server(debug=False, host=host, port=port)
```

```python
# get number of related bills passed by each senator
# can be used to show node size
def get_num_bills_filed(matrix):
    return np.sum(matrix, axis = 1)
    
get_num_bills_filed(matrix)
```

```python

def get_edge_trace(G):
    # given the graph G, get edge trace (x and y coords of links)
    edge_trace = {"x":[], "y":[], "weight":[]}
    
    for edge in G.edges(): 
        x0, y0 = G.node[edge[0]]['pos']
        x1, y1 = G.node[edge[1]]['pos']
        # thickness based on number of connections
        weight = G.edges[edge[0], edge[1]]["weight"]
        edge_trace['x'] += tuple([x0, x1, None]) # none to break segment
        edge_trace['y'] += tuple([y0, y1, None]) # none to break segment
        edge_trace["weight"] += [weight]
    return edge_trace

def get_node_trace(G):
    # given the graph G, get edge trace (x and y coords of links)
    node_trace = {"x":[], "y":[]}
    for node in G.nodes():
        x, y = G.node[node]['pos']
        node_trace['x'] += tuple([x])
        node_trace['y'] += tuple([y])
    return node_trace

def get_degrees(G):
    deg = [v for k, v in G.degree]
    return deg
```

```python
# create dropdown options
def create_options(unique_pri_com):
    _ = [{"label": pc, "value": "pc_"+pc} for pc in unique_pri_com] + [{"label": "All", "value": "all"}]
    return _

create_options(unique_pri_com)
```

```python
# create a list of colors with alphas for line based on edge weight
def make_line_colors(G):
    weights = get_edge_trace(G)['weight']
    weights2 = (weights - np.min(weights)) / (np.max(weights) - np.min(weights))
    colors = {(f'rgba(22, 96, 167, {i})') for i in weights2}
    return colors
make_line_colors(G)
```
