# Analysis of Senators Network Through Co-Filed Bills

## Executive Summary

Data of Philippine senators and bills filed from June 30, 2016 to February 4, 2019 were scraped from www.senate.gov.ph. The data was cleaned and organized for network analysis using NetworkX. The nodes are senators and senators are linked if they coauthored a bill. Considering all bills, the network of senators is almost fully connected with average degree per node of 23.6 and clustering coefficient of 0.99. Cayetano had the lowest degree at 18. An interactive dashboard was developed that can filter the bills based on the primary committee. The dashboard was used for further analysis of networks.

Below are some insights from exploring the senatorial bill co-filing network on different primary committees:
* For education, arts, and culture, Aquino filed the most number of bills at 99 bills and co-filed with all other senators at least once (degree = 23)
* For science and technology, Aquino filed the most bills at 38 and also linked with the most senators: 14.
* For education, Aquino filed the most bills, and had highest links (23)
* For justice and human rights, Gordon filed the most bills (63) with 19 links, while Legarda had highest links (20).
* For banks, financial institutions and currencies, Escudero filed the most bills (17) with 11 links.
* For economic affairs, Gatchalian filed the most bills (16) and highest number of links (8)

## About the Data

The data was scraped from www.senate.gov.ph. The data ran from June 30, 2016 to February 4, 2019. The items scraped are:
1. Bill filing date
2. Proponents or authors
3. Senate bill number
4. Short title
5. Long title
6. Scope whether local or national
7. Status, e.g. pending second reading, pending the committee, signed by the President
8. Subject
9. Primary committee
10. Secondary committee

## Network Analysis

```python
import networkx as nx
import plotly.plotly as py
import plotly.graph_objs as go

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

The one-hot encoded dataset processed above was loaded.

```python
df1 = pd.read_csv("bills_data_processed_17th_congress.csv", delimiter=',')
```

We defined a function for selecting the senators corresponding to a specific query for primary committee (or secondary committee.)

```python
def filter_return_props(df, filter_string):
    if filter_string == "all":
            df_filtered = df[:]
    else:
        df_filtered = df[df[filter_string] == 1]

    props = df_filtered.proponent.values
    props = [props[i] for i in range(len(props))] # convert to list
    return props
```

The function below selects the unique senators for the filtered dataframe.

```python
import ast

def unique_elements(element_set, truncate=False):
    if truncate == True:
        unique_elems = list(set([item.split(',')[0].replace(",", "") for sublist in element_set for item in ast.literal_eval(sublist)]))
    else:
        unique_elems = list(set([item for sublist in element_set for item in ast.literal_eval(sublist)]))
    return unique_elems
```

The function below makes an adjacency matrix given the list of senators per bill and the unique senators.

```python
def make_matrix(element_set, unique_elements, include_diag=False):
    # making adjacency matrix based on processed dataset [[a, v], [a, c]]

    matrix = np.zeros((len(unique_elements), len(unique_elements)))

    for i in range(len(unique_elements)):
        for j in range(len(unique_elements)):
            for x in element_set:
                if include_diag == True:
                    # including filed by same 1 proppnent
                    if i == j:
                        x = ast.literal_eval(x)
                        if len(x) == 1 and unique_elements[i] in x and unique_elements[j] in x:
                            matrix[i][j] += 1
                    if i != j:                
                        if unique_elements[i] in x and unique_elements[j] in x:
                            matrix[i][j] += 1
                else:
                    if i != j:                
                        if unique_elements[i] in x and unique_elements[j] in x:
                            matrix[i][j] += 1
    return matrix
```

We check the adjacency matrix of senators who co-filed a bill. regardless of primary committee.

```python
elems = filter_return_props(df1, "all")
unique_elems = unique_elements(elems)
matrix = make_matrix(elems, unique_elems, include_diag = False)
matrix.shape
```

The function below returns a NetworkX graph given an adjacency matrix. The second function plots the graph given a networkX graph G. And the third function returns the properties: average degree, number of edges, number of nodes, and average clustering coefficient.


```python
from collections import Counter
def get_G(matrix):
    mapping={a: b for a, b in zip(range(len(unique_elems)), unique_elems)}
    G = nx.from_numpy_matrix(matrix)
    G = nx.relabel_nodes(G,mapping)
    pos = nx.spring_layout(G);
#     given G networkX graph, and networkx position, pos, set position as nodes attribute
    nx.set_node_attributes(G, pos, 'pos')
    return G

def plot_graph(G):
    # plot network
    pos = nx.get_node_attributes(G,'pos')
    edges = G.edges
    weights = [G[u][v]['weight'] for u,v in edges]
    
    width = 2 * (np.array(weights) + 1 - np.min(np.array(weights)))/(np.max(weights) - np.min((weights)))

    plt.figure(figsize=(15, 15))
    nx.draw_networkx_nodes(G, pos, node_size=100, node_color="red");
    nx.draw_networkx_edges(G, pos, alpha=0.750, width=width);
    nx.draw_networkx_labels(G, pos, font_color="blue", font_size=10);
    plt.title("Senator co-filing of bill network from Jun 30, 2016 to Feb 4, 2019")
    plt.margins(0.2)
    
def show_properties(G):
    # Properties
    print("num edges: ", G.number_of_edges())
    print("num nodes: ", G.number_of_nodes())
    print("average clustering coef: ", nx.average_clustering(G))
    print("average degree: ", np.mean([v for k, v in G.degree]))
    
    
def get_hist_vals(G):
    # plot degree histogram
    # no. of nodes per degree
    degs = G.degree
    degs = [j for i, j in degs][2:]
    x = np.array([x for x, y in Counter(degs).items()])
    inds = np.argsort(x)
    x = x[inds]
    y = np.array([y for x, y in Counter(degs).items()])[inds]
    
    avg_deg = np.mean(degs)
    
    return x, y, avg_deg
    
```

### All bills

```python
primary_com = "all"
elems = filter_return_props(df1, primary_com)
unique_elems = unique_elements(elems)
matrix = make_matrix(elems, unique_elems, include_diag = False)
matrix.shape

G = get_G(matrix)
plot_graph(G)
```

## Interactive Dashboard

```python
from IPython.display import display, IFrame, HTML
import os


def show_app(app, port=9999, width=900, height=700):
    '''
    Accept app - a dash.Dash() data
    Return a running dash data running on the server
    '''

    # we set the server (host)
    host = 'localhost'

    # we set the URL of the page as localhost:9999
    url = f'http://{host}:{port}'  # use f to pass the variable in {}

    # display in HTML an open in new tab link
    display(HTML(f"<a href='{url}' target='_blank'>Open in a new tab </a>"))

    # display url in IFrame
    display(IFrame(url, width=width, height=height))

    # ???
    app.css.config.serve_locally = True
    app.scripts.config.serve_locally = True

    # Return app = dash.Dash() running server
    return app.run_server(debug=False, host=host, port=port)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
from dash.dependencies import Input, Output, State, Event
import pandas as pd
```

```python
# begin dash object framework thingy
app = dash.Dash()
```

```python
# to load CSS and other static stuff
app = dash.Dash(static_folder='assets/')
```

```python
# get number of related bills passed by each senator
# can be used to show node size
def get_num_bills_filed(matrix):
    return np.sum(matrix, axis = 1)
```

```python
def get_edge_trace(G):
    # given the graph G, get edge trace (x and y coords of links)
    edge_trace = {"x":[], "y":[], "weight":[]}
    
    for edge in G.edges(): 
        x0, y0 = G.node[edge[0]]['pos']
        x1, y1 = G.node[edge[1]]['pos']
        # thickness based on number of connections
        weight = G.edges[edge[0], edge[1]]["weight"]
        edge_trace['x'] += [tuple([x0, x1])] # none to break segment
        edge_trace['y'] += [tuple([y0, y1])] # none to break segment
        edge_trace["weight"] += [weight]
    return edge_trace

def get_node_trace(G):
    # given the graph G, get edge trace (x and y coords of links)
    node_trace = {"x":[], "y":[], "name":[]}
    for node in G.nodes():
        x, y = G.node[node]['pos']
        node_trace['x'] += tuple([x])
        node_trace['y'] += tuple([y])
        node_trace['name'] += [node]
    return node_trace

def get_degrees(G):
    deg = [v for k, v in G.degree]
    return deg
```

```python
# create dropdown options
def create_options(unique_pri_com):
    _ = [{"label": pc, "value": "pc_"+pc} for pc in unique_pri_com] + [{"label": "All", "value": "all"}]
    return _
```

```python
# create a list of colors with alphas for line based on edge weight
def make_line_colors(G):
    weights = get_edge_trace(G)['weight']
    weights2 = (weights - np.min(weights)) / (np.max(weights) - np.min(weights))
    colors = {(f'rgba(22, 96, 167, {i})') for i in weights2}
    return colors
```

```python
# initialize
primary_com = "all"
elems = filter_return_props(df1, primary_com)
unique_elems = unique_elements(elems)
matrix = make_matrix(elems, unique_elems, include_diag = False)

G = get_G(matrix)

edges_x = get_edge_trace(G)['x']
edges_y = get_edge_trace(G)['y']
edges_weight = np.array(get_edge_trace(G)['weight'])
names = [i.split(", ")[0] for i in np.array(G.nodes)] # for display

# rescale edges_weight
edges_weight = 2 * (edges_weight+1 - np.min(edges_weight))/(np.max(edges_weight) - np.min(edges_weight))


# dashboard

app.layout = html.Div([
    html.H1('Senator Network', style={'textAlign': 'center'}),
    html.P("Network of senators with connections based on bills co-authored", 
           style={'textAlign': 'center'}),
    
    html.Label('First committee (subject)', style={'textAlign': 'center'}),
    
    # THIS IS THE DROPDOWN FOR TYPE OF TOPIC (COMMITTEE)
    dcc.Dropdown(
        options= create_options(unique_pri_com), # label presented to user, value to backend

        # able to select multiple values?
        multi=False,

        # preselect values
        value='all',
        style={'textAlign': 'center'},
        id = "dropdown",
        ),
    
    
    # Plotting the nodes (scatter) and links (lines)
    # Size based on number of bills passed?
    # Size based on degree?
    
    dcc.Graph(figure={
        'data': [go.Scatter(x=edges_x[i], y=edges_y[i], mode='lines', 
                        line = dict(width = edges_weight[i], color='#4B8BBE'), name="links", hoverinfo="text")\
                 for i in range(len(edges_x))] +\
            
            [go.Scatter(x=get_node_trace(G)['x'], y=get_node_trace(G)['y'], mode='markers+text', 
                       name="size: number of bills filed", text = names, 
                       textposition='top center', hoverinfo="text", 
                       hovertext = list(zip(names, ["degree: " + str(i) for i in get_degrees(G)], 
                                       ["bills filed: "+ str(i) for i in get_num_bills_filed(make_matrix(elems, unique_elems, include_diag = True))])),
                        # No. of degrees marker
                       # marker = dict(size = 20 * (get_degrees(G) - np.min(get_degrees(G))) / (np.max(get_degrees(G)) - np.min(get_degrees(G))))),
                       # No. of bills passed marker
                       marker = dict(color = "orange",size = 20 * (get_num_bills_filed(make_matrix(elems, unique_elems, include_diag = True)) -\
                                                  np.min(get_num_bills_filed(make_matrix(elems, unique_elems, include_diag = True)))) / \
                                     (np.max(get_num_bills_filed(make_matrix(elems, unique_elems, include_diag = True))) - \
                                      np.min(get_num_bills_filed(make_matrix(elems, unique_elems, include_diag = True)))))),
    
        ],
        'layout': {
            
            'showlegend':False,
            
            'title':'Senator network by co-filing of bills for all bills',
             'xaxis':{
                 "showgrid":False,
                 "zeroline": False,
                  "showticklabels" :False
             },
            'yaxis':{
                 "showgrid":False,
                 "zeroline": False,
                "showticklabels" :False
             },
            "height": 600,
            "width": 1000,
        }
    }
              , style={'textAlign': 'center'}, id='network'),
    
    # NETWORK PROPERTIES
    html.Span([html.H6("Network Properties", style={'textAlign': 'center'})]),
    html.Div(
        [html.P("Avg degree: {}".format(np.round(np.mean(get_degrees(G)), 2))),
        html.P("Avg clustering coefficient: {}".format(np.round(nx.average_clustering(G), 2))),
        html.P("Num edges: {}".format(G.number_of_edges())),
            html.P("Num nodes: {}".format(G.number_of_nodes()))], 
        style={'textAlign': 'center'},
        id="properties"),
    
    
    html.Br(),
    html.Br(),
    html.P(["The data was scraped from www.senate.gov.ph. The data are all bills filed by senators from June 30, 2016 to February 4, 2019."]),
    
    html.Br(),
    html.Br()
    
    
])

app.callback_map = {} # resets callbacks

# Callbacks
@app.callback(Output('network', 'figure'), [Input('dropdown', 'value')])
def update_graph(dropdown_value):
    
    elems = filter_return_props(df1, dropdown_value)
    unique_elems = unique_elements(elems, truncate=False) # for matrix 1 and 2
    matrix = make_matrix(elems, unique_elems) # for links
    matrix2 = make_matrix(elems, unique_elems, include_diag=True) # for number of bills
    
    X = get_G(matrix)
    pos = nx.get_node_attributes(X,'pos')
    edges_x = get_edge_trace(X)['x']
    edges_y = get_edge_trace(X)['y']
    nodes_x = get_node_trace(X)['x']
    nodes_y = get_node_trace(X)['y']
    
    edges_weight = np.array(get_edge_trace(X)['weight'])
    
    names = [i.split(", ")[0] for i in unique_elems] # for display
    degrees = get_degrees(X)

    
    # rescale edges_weight
    edges_weight = 2 * (edges_weight+1 - np.min(edges_weight))/(np.max(edges_weight) - np.min(edges_weight))
    
    new_fig = {
        'data': [go.Scatter(x=edges_x[i], y=edges_y[i], mode='lines', 
                        line = dict(width = edges_weight[i], color='#4B8BBE'), name="links", hoverinfo="text")\
                 for i in range(len(edges_x))] + \
        
                # alpha
                [go.Scatter(x=nodes_x, y=nodes_y, mode='markers+text', 
                       name="size: number of bills filed", text = names, 
                       textposition='top center', hoverinfo="text",
                       hovertext = list(zip(names,
                                        ["degree: " + str(i) for i in degrees], 
                                       ["bills filed: "+ str(i) for i in get_num_bills_filed(matrix2)])),
                       
                        # No. of bills passed marker
                       marker = dict(color = "orange", size = 20 * (get_num_bills_filed(matrix2) -\
                                                  np.min(get_num_bills_filed(matrix2))) / \
                                     (np.max(get_num_bills_filed(matrix2)) - \
                                      np.min(get_num_bills_filed(matrix2)))))],
        'layout': {
            
            'showlegend':False,
            
            'title':'Senator network by co-filing of bills for {}'.format(dropdown_value),
             'xaxis':{
                 "showgrid":False,
                 "zeroline": False,
                  "showticklabels" :False
             },
            'yaxis':{
                 "showgrid":False,
                 "zeroline": False,
                "showticklabels" :False
             },
            "height": 600,
            "width": 1000,
        }
    }
    
    return new_fig

# for average degree
@app.callback(Output('properties', 'children'), [Input('dropdown', 'value')])
def update_properties(dropdown_value):
    elems = filter_return_props(df1, dropdown_value)
    unique_elems = unique_elements(elems, truncate=True)
    matrix = make_matrix(elems, unique_elems)
    
    X = get_G(matrix)
    names = [i.split(", ")[0] for i in list(X.nodes)] # for display
    
    value = [html.P("Avg degree: {}".format(np.round(np.mean(get_degrees(X)), 2))),
        html.P("Avg clustering coefficient: {}".format(np.round(nx.average_clustering(X), 2))),
        html.P("Num edges: {}".format(X.number_of_edges())),
            html.P("Num nodes: {}".format(X.number_of_nodes()))]
    return value


# NOTES 
# Rappler article
# Deploy stuff
# Consistency with stance/advocacies during elections
# Hubs - if you remove what will happen
# number of bills over time
# Status of bill
```
